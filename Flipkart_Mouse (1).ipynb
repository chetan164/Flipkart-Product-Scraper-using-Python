```python
import requests # it is used to fetch the html content from a website
from bs4 import BeautifulSoup  # it is used to find specific data from a html code
import time
```

```python
url='https://www.flipkart.com/search?q=phones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/115.0.0.0 Safari/537.36"
    )
}


response=requests.get(url,headers=headers)
response
```

```python
soup=BeautifulSoup(response.content,"html.parser")
soup
```

```python
# create a list of all the products
name_list=[]
elements=soup.find_all("div",class_='RG5Slk')
for i in range(len(elements)):
    name_list.append(elements[i].get_text().strip())

name_list
```

```python
# create a list of products price
price_list=[]
elements=soup.find_all("div",class_='hZ3P6w DeU9vF')
for i in range(len(elements)):
    price_list.append(elements[i].get_text().strip())

price_list
```

```python
# create a list of discount and store it in a new list
discount_list=[]
elements=soup.find_all("div",class_='HQe8jr')
for i in range(len(elements)):
    discount_list.append(elements[i].get_text().strip())

discount_list
```

```python
# ₹14,999  -> 14999
# 32% off  -> 32
# data Cleaning 
# re.sub ("₹,","")
```

```python
price_list
```

```python
import re
```

```python

=[]
for i in price_list:
    clean_price.append(re.sub("[₹,]","",i))

clean_price
```

```python
clean_discount=[]
for i in discount_list:
    clean_discount.append(re.sub("[%off]","",i))

clean_discount
```
